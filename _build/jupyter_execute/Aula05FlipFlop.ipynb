## Flip-flops e registradores

### Introdução
Circuitos digitais são interligações de diversas portas lógicas, que processam os dados no forma de sinais digitais, projetados para realizar operações aritméticas e lógicas.\
Os circuitos digitais são classificados em dois tipos
1. Circuitos combinacionais e
2. Circuitos sequenciais.
   


## Circuitos Combinacionais

Os circuitos combinacionais são circuitos independentes do tempo, que dependem da entrada atual e não depende de entradas anteriores para gerar qualquer saída. A Figura (1) mostra o diagrama de blocos
de circuito combinacional.



As saídas do circuito combinacional dependem apenas das entradas presentes. ou seja,

Características dos circuitos combinacionais:
1. Simples na construção
2. A velocidade é rápida, ocorre apenas atraso de propagação.
3. Sem feedback
4. Os blocos de construção básicos são AND gate, OR gate e NOT gate.
5. Independente do tempo
   
Exemplos: Somadores, Subtratores, Decodificadores, Codificadores, Multiplexadores, Demultiplexadores etc.

## Circuitos sequenciais

Circuitos sequenciais são aqueles que dependem do tempo (ciclos de clock) e dependem também das saídas anteriores para gerar qualquer saída. A Figura (2) mostra o diagrama de blocos de circuito sequencial.



Pela figura (2), a saída não depende apenas das entradas atuais, mas também das saídas anteriores.

Os circuitos sequenciais enviam a saída anterior para a entrada do circuito com sinal feedback através do elemento de memória. O elemento de memória é usado para armazenar os dados anteriores e enviá-los ao circuito combinacional como entradas adicionais.

O elemento de memória desempenha um papel muito importante no projeto de circuitos sequenciais. \
flop/Latch é um bloco de construção básico/bloco de construção elementar que atua como um elemento de memória
para o projeto de quaisquer circuitos sequenciais. A presente seção discute o conceito de Flip-flops.
Os circuitos sequenciais são classificados em dois tipos,
1. Circuitos sequenciais síncronos – Esses circuitos geram a saída apenas em
instantes de tempo. Esses circuitos requerem pulso de clock (gerador de clock mestre) para amostrar as entradas e
determinar o comportamento do sistema. Esses circuitos também são chamados de circuitos sequenciais com clock.
2. Circuitos sequenciais assíncronos – Esses circuitos geram a saída instantaneamente, uma vez
as variáveis ​​de entrada mudam. Esses circuitos também são chamados de circuitos sequenciais sem clock.

Características dos circuitos sequenciais:
1. A resultado depende tanto dos dados presentes como dos passados.
2. O feedback está presente para tornar dependente dos resultados anteriores/passados.
3. A velocidade é menor que os circuitos combinacionais.
4. Projeto complexo que os circuitos combinacionais.
5. Depende do tempo, portanto precisa de relógio para acionamento.
6. Projetado para armazenar os dados.
7. O elemento básico é o flip-flop.
Exemplo: Contadores, Registradores etc...

### Conceito de Flip-Flops
O bloco de construção elementar de qualquer circuito sequencial é o flip-flop, o próprio flip-flop é
um circuito sequencial simples. Basicamente, os circuitos sequenciais possuem feedback, e vê-se que o feedback
também está presente nos flip-flops.\

Os flip-flops têm duas condições estáveis; cada condição estável é chamada de estado.\
O Estado representa o armazenamento de símbolos binários. O estado de um sistema também é chamado de conteúdo ou estado interno.

O flip-flop possui dois estados; portanto, os flip-flops são chamados de dispositivos/elementos biestáveis.

## Latches:
Latches:
As Latches são uma classe de flip-flops; o tempo das mudanças de saída não é controlado em Latches. A saída responde imediatamente às alterações nas linhas de entrada. Portanto, as linhas de entrada são precisam ser continuamente interrogados. Nesta seção, Latches SR usando realização NOR e NAND realização são discutidos.

1. Latches SR:

Nos flip-flops, o armazenamento de '1' é chamado de Set e o armazenamento de '0' é chamado de Reset, daí o nome latch SR.
ou seja, definir e redefinir a trava. Figura (4) Mostra a realização da porta NOR do latch SR.



Na figura, $S$ e $R$ são entradas Set e Reset respectivamente, $G_1$ e $G_2$ são portas NOR, $Q$ e $\overline{Q}$ são linhas de saída.

O circuito pode ser analisado pelos seguintes casos:

_Caso (i): Se $S=0$ e $R=1$_

Como sabemos, a saída da porta NOR é sempre '0' se qualquer das entradas é '1'. Então se $R=1$, então a saída de $G_1$ será '0' independente da outra entrada, ou seja $Q=0$ e a saída de $G_1$ é uma das entradas de $G_2$. Agora $G_2$ recebe os valores $S=0$ e $Q=0$, portanto $\overline{Q}=1$. Esta operação é chamada _reset_.

Com o mesmo estado de _reset_, consideremos $S=0$ e $R=0$. Este valor de $R$ e o valor de $\overline{Q}=1$ são as entradas para $G_1$, cuja saída resulta em $Q=0$ e, de modo similar, $S=0$ e $Q=0$ são as entradas para $G_2$, o que resulta em $\overline{Q}=1$. Da discussão anterior, para os valores de $R=S=0$ o latch SR retém o valor anterior. Esta é uma operação de memória - armazenamento do símbolo binário '0'.

_Caso (ii): Se $S=1$ e $R=0$_

Se $S=1$, a saída de $G_2$ é '0' independente da outra entrada, ou seja, $\overline{Q}=0$ e a saída de $G_2$ é uma das entradas para $G_1$. Agora, $G_1$ toma como entrada os valores $R=0$ e $\overline{Q}=0$, portanto $Q=1$. Esta operação é denominada _set_.

Consideremos agora $S=0$ e $R=0$ sob o estado de _set_. $S=0$ e a saída anterior $Q=1$ são as entradas de $G_2$, cuja saída é $\overline{Q}=0$ e consequentemente $R=0$ e $\overline{Q}=0$ são as entradas de $G_1$, gerando a saída $Q=1$. Desta discussão, para os valores de $R=S=0$, o latch SR mantém o dado anterior. Esta é uma operação de memória - armazenamento do símbolo binário '1'.

_Caso (iii): Se $S=1$ e $R=1$_

Se $S=1$, independente da outra entrada, a saída de $G_2$ será $\overline{Q}=0$. $G_1$ toma as entradas $R=1$ e $\overline{Q}=0$ resultando em $Q=0$. Este é um caso inválido/não usado/proibido do latch SR.

3. Latches SR síncronos

Em latches SR e $\overline{SR}$ as saídas serão modificadas imediatamente após a mudança nos valores de entrada. É frequentemente desejável evitar que as mudanças dos valores de entrada afetem imediatamente os valores de saída. Para permitir que as mudanças na entrada sejam efetivas apenas dentro de um período de tempo específico, um sinal de habilitação é introduzido. Latches SR com este novo sinal são denominados latches SR síncronos ou latches temporizados, ou latches controlados.

As entradas do latch $\overline{SR}$ são denotadas $S^*$ e $R^*$, e estes valores dependem de $S$, $R$ e do sinal de habilitação $En$ (_enable_).

O latch SR síncrono acima pode ser analizado com base nos casos a seguir. Antes, tomemos nota das expressões para $S^*$ e $R^*$

$S^* = \overline{S.En}=\overline{S}+\overline{En}\qquad(4)$

$R^* = \overline{R.En}=\overline{R}+\overline{En}\qquad(5)$

_Caso (i): $S=X$, $R=X$ e $En=0$_

Se $En=0$, independente dos valores de $S$ e $R$, $S^*=R^*=1$ e para estes valores o latch $\overline{SR}$ apresenta como saída o próprio estado anterior como operação de memória. Este caso evidencia a característica do latch SR síncrono de ser insensível aos valores de entrada devido ao sinal de controle.

_Caso (ii): $S=0$, $R=0$ e $En=1$_

Se $S=R=0$ e $En=1$, então $S^*=1$ e $R^*=1$ o que resulta em operação de memória.

_Caso (iii): $S=0$, $R=1$ e $En=1$_

Se $S=0$, $R=1$ e $En=1$, então $S^*=1$ e $R^*=0$ o que resulta em operação de _reset_ - armazenamento do símbolo binário '0'.

_Caso (iv): $S=1$, $R=0$ e $En=1$_

Se $S=1$, $R=0$ e $En=1$, então $S^*=0$ e $R^*=1$ o que resulta em operação de _set_ - armazenamento do símbolo binário '1'.

_Caso (v): $S=1$, $R=1$ e $En=1$_

Se $S=1$, $R=1$ e $En=1$, então $S^*=1$ e $R^*=1$ o que resulta em estado inválido.

5. Diferenças entre latch e flip-flop

| Latch | Flip-flop |
| :-: | :-: |
| O sinal de controle não é um _clock_ | O sinal de controle é um _clock_ |
| Sensível a nível | Sensível a borda |
| Ativado por nível | Ativado por pulso e por borda |
| Resposta imediata da saída dentro dos delays de propagação | Mudanças na saída ocorrem de acordo com mudanças em um sinal de controle |

6. Sinal de _clock_ ou pulso de _clock_

Pulso de _clock_ é uma mudança de amplitude (voltagem) precisamente espaçada e contínua. Este sinal determina o tempo da aplicação das entradas sobre o sistema. De modo simplificado, pulsos de _clock_ têm a forma de onda quadrada com 50% de ciclo de trabalho (_duty cycle_). A figura abaixo ilustra um sinal de _clock_ com 50% de _duty cicle_ que é utilizado para sincronizar as operações de circuitos digitais sequenciais.



Os parâmetros característicos de um sinal de _clock_ são sua frequência ($f$) e período ($T$). Frequência descreve a quantidade de ciclos de saída executados dentro de um segundo, enquanto o período é o tempo total necessário para que o circuito complete um ciclo completo. Período é o inverso da frequência, ou seja:

$T = \frac{1}{f}\qquad(8)$

com $T$ em segundos e consequentemente $f$ em Hertz (_Hz_).

O _duty cycle_ de um sinal de _clock_ é a proporção de tempo do sinal do estado ON relativo ao tempo total. A figura abaixo ilustra os períodos de operação para referência.



$T_{Total} = T_{ON}+T_{OFF}\qquad(9)$

_Duty Cycle_$\ (\%) = \frac{T_{ON}}{T_{Total}}*100\qquad(10)$

7. Considerações de _timing_

A resposta de qualquer sistema real é não-instantânea relativa à aplicação das entradas, ou seja, os sistemas irão levar determinado tempo para responder às entradas. Estes atrasos ocorrem em função dos atrasos intrínsecos às próprias portas lógicas utilizadas. As limitações dos dispositivos empregados devem ser levadas em consideração para obtenção de resposta dentro de um período alvo.

O tempo entre a obtenção de uma resposta na saída após a mudança de uma entrada é denominado atraso (_delay_) de propagação. O atraso de propagação entre cada par de terminais de entrada e saída é diferente devido às mudanças de estado de baixo para alto ou vice-versa.

Tomemos os as operações de _set_ e _reset_ realizadas sobre um latch SR como exemplo. Também é importante observar a resposta do latch SR no instante $t_{15}$, onde ambos os sinais de entrada mudam simultaneamente de '1' para '0'. Como consequência a resposta do latch SR é imprevisível, pois este pode estar no estado de '0' ou metaestável. A aplicação de $S=1$ no instante $t_{16}$ retorna o latch ao seu comportamento previsível após um breve atraso de propagação.
