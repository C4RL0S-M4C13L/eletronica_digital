# DECODIFICADORES x CODIFICADORES
## Decodificador:
### 1. Circuito digital que faz a conversão de um código binário para outro código ou um número qualquer;
### 2. Geralmente recebe um código binário na entrada e ativa apenas 1 saída, correspondente ao número decodificado;

## Codificador:
### 1. Circuito digital que faz a conversão de um número (ou um código qualquer) para um código binário;
### 2. Geralmente recebe um dado de entrada com M bits e tem como saída um código de N bits;
